```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import gmaps
from api_keys import g_key
from ipywidgets.embed import embed_minimal_html
```

```python
montreal_collisions = pd.read_csv('Resources/montreal_2012_2018.csv')
montreal_incomes = pd.read_csv('Resources/montreal_incomes.csv')
```

```python
montreal_collisions=montreal_collisions.rename(columns={'LOC_LONG':'Longitude',
                                                        'LOC_LAT':'Latitude',
                                                        'AN': 'Year',
                                                        'DT_ACCDN': 'Date',
                                                        'HR_ACCDN':'Hour',
                                                        'JR_SEMN_ACCDN':'Weekday',
                                                        'CD_COND_METEO':'Weather Conditions',
                                                        'CD_MUNCP':'District'
                                                       })
```

```python
montreal_collisions= montreal_collisions.loc[montreal_collisions["GRAVITE"]== "Mortel"]
montreal_collisions= montreal_collisions.loc[montreal_collisions["REG_ADM"]== "Montréal(06)"]
```

```python
montreal_districts = pd.read_csv('Resources/montreal_districts.csv')
```

```python
montreal_collisions=pd.merge(montreal_collisions, montreal_districts, on= "District")
```

```python
montreal_collisions['Hour'] = montreal_collisions['Hour'].str[:2]
```

```python
montreal_collisions['Hour']=montreal_collisions['Hour'].replace('No',99)
```

```python
montreal_collisions['Hour']=montreal_collisions['Hour'].astype(int)
```

```python
montreal_collision=montreal_collisions[['Longitude', 'Latitude', 'Year','Date','Hour','Districts']]
montreal_collision
```

```python
montreal_collision=montreal_collision.sort_values('Year',ascending=False)
montreal_collision.dtypes
```

```python
montreal_collision['Time Patterns']=0
```

```python
montreal_collision['Time Patterns'][montreal_collision['Hour'].isin(range(0,5))]='Early Morning'
montreal_collision['Time Patterns'][montreal_collision['Hour'].isin(range(4,10))]='Morning Commute'
montreal_collision['Time Patterns'][montreal_collision['Hour'].isin(range(9,18))]='Working Hours'
montreal_collision['Time Patterns'][montreal_collision['Hour'].isin(range(17,21))]='Evening Commute'
montreal_collision['Time Patterns'][montreal_collision['Hour'].isin(range(20,25))]='Evening'
```
